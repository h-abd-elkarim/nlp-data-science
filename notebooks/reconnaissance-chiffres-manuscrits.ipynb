```python
from sklearn.datasets import fetch_openml
import matplotlib.pyplot as plt
```

```python
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
```

```python
# diviser la table
data = mnist.data # les image
labels = mnist.target.astype(int)  # les chiffres
```

```python
print("5 premières images sous forme de vecteurs :")
print(data[:5])
print("\nChiffres correspondantes :")
print(labels[:10])
```

```python
# Afficher la dimensionnalité des données
print("\nDimensions de data :", data.shape)
print("Dimensions de labels :", labels.shape)
```

```python
import random

index = random.randint(0, len(data) - 1)

plt.imshow(data[index].reshape(28, 28), cmap='gray')
plt.title(f"Number : {labels[index]}")
plt.axis('off')
plt.show()

```
