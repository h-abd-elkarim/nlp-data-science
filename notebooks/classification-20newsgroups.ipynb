```python
import gzip
import xml.etree.ElementTree as ET
import spacy
import nltk
from nltk.util import ngrams
from collections import Counter
from nltk import Tree
import math

# Télécharger les ressources NLTK nécessaires
nltk.download('punkt')  # Assurez-vous de télécharger 'punkt' pour la tokenisation
nltk.download('punkt_tab')
# Charger le modèle spaCy
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    import spacy.cli
    spacy.cli.download("en_core_web_sm")
    nlp = spacy.load("en_core_web_sm")

def decompress_gz(xml_file):
    """Décompresse un fichier .gz."""
    if xml_file.endswith(".gz"):
        print("Décompression du fichier en cours...")
        with gzip.open(xml_file, 'rb') as f_in:
            decompressed_path = xml_file[:-3]
            with open(decompressed_path, 'wb') as f_out:
                f_out.write(f_in.read())
        print(f"Fichier décompressé : {decompressed_path}")
        return decompressed_path
    return xml_file

def extract_abstracts(xml_file, limit=15):
    """Extrait les 15 premiers abstracts du fichier XML."""
    tree = ET.parse(xml_file)
    root = tree.getroot()

    abstracts = []
    count = 0
    for abstract in root.iter("AbstractText"):
        if abstract.text:
            abstracts.append(abstract.text.strip())
            count += 1
        if count >= limit:
            break

    return abstracts

def generate_parse_tree(sentence):
    """Génère l'arbre syntaxique d'une phrase."""
    
    def token_to_nltk_tree(token):
        """Convertit un token spaCy en sous-arbre nltk."""
        if token.n_lefts + token.n_rights > 0:
            return Tree(token.orth_, [token_to_nltk_tree(child) for child in token.children])
        else:
            return token.orth_

    # Vérifier si la phrase contient au moins un mot
    if sentence and sentence.root:
        return token_to_nltk_tree(sentence.root)
    else:
        return None

def build_ngram_model(text, n=2):
    """Construit un modèle N-gram basé sur un texte donné."""
    # Tokenisation avec NLTK
    tokens = nltk.word_tokenize(text.lower())  # Tokenisation et mise en minuscules
    n_grams = list(ngrams(tokens, n))  # Créer des n-grams
    ngram_freq = Counter(n_grams)  # Fréquence des n-grams
    unigram_freq = Counter([gram[0] for gram in n_grams])  # Fréquence des unigrams
    return ngram_freq, unigram_freq, tokens

def calculate_bigram_probability(bigram, ngram_freq, unigram_freq):
    """Calcule la probabilité d'un bigramme."""
    first_word = bigram[0]
    bigram_count = ngram_freq[bigram]
    unigram_count = unigram_freq[first_word]
    
    # La probabilité P(w2|w1) = C(w1, w2) / C(w1)
    if unigram_count > 0:
        return bigram_count / unigram_count
    else:
        return 0

def calculate_perplexity(text, ngram_freq, unigram_freq, n=2):
    """Calcule la perplexité d'un texte donné basé sur le modèle N-gram."""
    tokens = nltk.word_tokenize(text.lower())  # Tokenisation du texte
    bigram_list = list(ngrams(tokens, n))  # Créer des bigrams

    log_prob = 0
    for bigram in bigram_list:
        prob = calculate_bigram_probability(bigram, ngram_freq, unigram_freq)
        if prob > 0:
            log_prob += math.log(prob)
        else:
            # Si la probabilité est 0, cela signifie que le bigramme n'a pas été vu. On l'ignore.
            log_prob += math.log(1e-10)  # Utilisation d'un très petit nombre pour éviter les problèmes

    # Normalisation par le nombre de tokens (en minuscules) dans le texte
    perplexity = math.exp(-log_prob / len(bigram_list))
    return perplexity

def build_pos_transition_matrices(sentences):
    """Construit les matrices de transition de POS (unigramme et bigramme)."""
    unigram_pos_count = Counter()
    bigram_pos_count = Counter()
    
    # Parcours des phrases et POS tags
    for sentence in sentences:
        pos_tags = [token.pos_ for token in sentence]
        
        # Unigramme des POS tags
        unigram_pos_count.update(pos_tags)
        
        # Bigramme des POS tags
        bigram_pos_count.update(ngrams(pos_tags, 2))
    
    # Calcul des probabilités
    unigram_pos_prob = {pos: count / sum(unigram_pos_count.values()) for pos, count in unigram_pos_count.items()}
    bigram_pos_prob = {bigram: count / unigram_pos_count[bigram[0]] for bigram, count in bigram_pos_count.items()}
    
    return unigram_pos_prob, bigram_pos_prob

def build_ner_transition_matrices(sentences):
    """Construit les matrices de transition des entités nommées (unigramme et bigramme)."""
    unigram_ner_count = Counter()
    bigram_ner_count = Counter()
    
    # Parcours des phrases et des entités nommées
    for sentence in sentences:
        ner_tags = [ent.label_ for ent in sentence.ents]
        
        # Unigramme des NER tags
        unigram_ner_count.update(ner_tags)
        
        # Bigramme des NER tags
        bigram_ner_count.update(ngrams(ner_tags, 2))
    
    # Calcul des probabilités
    unigram_ner_prob = {ner: count / sum(unigram_ner_count.values()) for ner, count in unigram_ner_count.items()}
    bigram_ner_prob = {bigram: count / unigram_ner_count[bigram[0]] for bigram, count in bigram_ner_count.items()}
    
    return unigram_ner_prob, bigram_ner_prob

# Spécifie le chemin du fichier XML
xml_path = "C:/Users/user/Downloads/sample-0001.xml.gz"
xml_file = decompress_gz(xml_path)

# Extraire les 15 premiers abstracts
abstracts = extract_abstracts(xml_file, limit=15)

# Afficher les analyses morphosyntaxiques
print(f"Extraits des {len(abstracts)} premiers abstracts avec analyse morphosyntaxique :")

for i, abstract in enumerate(abstracts, 1):
    print(f"\nAbstract {i} : {abstract}")

    # Segmenter en phrases
    doc = nlp(abstract)
    sentences = list(doc.sents)

    # Calcul des matrices de transition POS
    unigram_pos_prob, bigram_pos_prob = build_pos_transition_matrices(sentences)
    
    # Affichage des matrices de transition
    print("\n--- Matrice de transition Unigramme POS ---")
    for pos, prob in unigram_pos_prob.items():
        print(f"{pos}: {prob}")
        
    print("\n--- Matrice de transition Bigramme POS ---")
    for bigram, prob in bigram_pos_prob.items():
        print(f"{bigram}: {prob}")

    # Calcul des matrices de transition NER
    unigram_ner_prob, bigram_ner_prob = build_ner_transition_matrices(sentences)
    
    # Affichage des matrices de transition NER
    print("\n--- Matrice de transition Unigramme NER ---")
    for ner, prob in unigram_ner_prob.items():
        print(f"{ner}: {prob}")
        
    print("\n--- Matrice de transition Bigramme NER ---")
    for bigram, prob in bigram_ner_prob.items():
        print(f"{bigram}: {prob}")

    # Afficher les analyses morphosyntaxiques
    for j, sentence in enumerate(sentences[:3], 1):
        print(f"\n--- Phrase {j} ---\n{sentence.text}")

        # Générer et afficher l'arbre syntaxique
        parse_tree = generate_parse_tree(sentence)
        if parse_tree:
            parse_tree.pretty_print()
        else:
            print("Arbre syntaxique non généré (phrase vide ou problème de parsing).")

    # Construire le modèle N-gram (bigram) pour chaque abstract
    bigram_model, unigram_model, tokens = build_ngram_model(abstract, n=2)
    
    # Afficher les fréquences des bigrammes
    print("\n--- Modèle N-gram (Bigramme) pour cet abstract ---")
    for bigram, freq in bigram_model.items():
        prob = calculate_bigram_probability(bigram, bigram_model, unigram_model)
        print(f"{bigram}: {freq}, Probabilité: {prob}")

    # Calculer et afficher la perplexité
    perplexity = calculate_perplexity(abstract, bigram_model, unigram_model, n=2)
    print(f"\nPerplexité pour l'abstract {i}: {perplexity}")
```

```python
import gzip
import xml.etree.ElementTree as ET
import spacy
import nltk
from nltk.util import ngrams
from collections import Counter
from nltk import Tree
import math
import gensim
from gensim import corpora
from nltk.corpus import stopwords

# Télécharger les ressources NLTK nécessaires
nltk.download('punkt')  # Assurez-vous de télécharger 'punkt' pour la tokenisation
nltk.download('stopwords')  # Télécharger les stopwords
nltk.download('punkt_tab')
# Charger le modèle spaCy
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    import spacy.cli
    spacy.cli.download("en_core_web_sm")
    nlp = spacy.load("en_core_web_sm")

def decompress_gz(xml_file):
    """Décompresse un fichier .gz."""
    if xml_file.endswith(".gz"):
        print("Décompression du fichier en cours...")
        with gzip.open(xml_file, 'rb') as f_in:
            decompressed_path = xml_file[:-3]
            with open(decompressed_path, 'wb') as f_out:
                f_out.write(f_in.read())
        print(f"Fichier décompressé : {decompressed_path}")
        return decompressed_path
    return xml_file

def extract_abstracts(xml_file, limit=15):
    """Extrait les 15 premiers abstracts du fichier XML."""
    tree = ET.parse(xml_file)
    root = tree.getroot()

    abstracts = []
    count = 0
    for abstract in root.iter("AbstractText"):
        if abstract.text:
            abstracts.append(abstract.text.strip())
            count += 1
        if count >= limit:
            break

    return abstracts

def generate_parse_tree(sentence):
    """Génère l'arbre syntaxique d'une phrase."""
    
    def token_to_nltk_tree(token):
        """Convertit un token spaCy en sous-arbre nltk."""
        if token.n_lefts + token.n_rights > 0:
            return Tree(token.orth_, [token_to_nltk_tree(child) for child in token.children])
        else:
            return token.orth_

    # Vérifier si la phrase contient au moins un mot
    if sentence and sentence.root:
        return token_to_nltk_tree(sentence.root)
    else:
        return None

def build_ngram_model(text, n=2):
    """Construit un modèle N-gram basé sur un texte donné."""
    # Tokenisation avec NLTK
    tokens = nltk.word_tokenize(text.lower())  # Tokenisation et mise en minuscules
    n_grams = list(ngrams(tokens, n))  # Créer des n-grams
    ngram_freq = Counter(n_grams)  # Fréquence des n-grams
    unigram_freq = Counter([gram[0] for gram in n_grams])  # Fréquence des unigrams
    return ngram_freq, unigram_freq, tokens

def calculate_bigram_probability(bigram, ngram_freq, unigram_freq):
    """Calcule la probabilité d'un bigramme."""
    first_word = bigram[0]
    bigram_count = ngram_freq[bigram]
    unigram_count = unigram_freq[first_word]
    
    # La probabilité P(w2|w1) = C(w1, w2) / C(w1)
    if unigram_count > 0:
        return bigram_count / unigram_count
    else:
        return 0

def calculate_perplexity(text, ngram_freq, unigram_freq, n=2):
    """Calcule la perplexité d'un texte donné basé sur le modèle N-gram."""
    tokens = nltk.word_tokenize(text.lower())  # Tokenisation du texte
    bigram_list = list(ngrams(tokens, n))  # Créer des bigrams

    log_prob = 0
    for bigram in bigram_list:
        prob = calculate_bigram_probability(bigram, ngram_freq, unigram_freq)
        if prob > 0:
            log_prob += math.log(prob)
        else:
            # Si la probabilité est 0, cela signifie que le bigramme n'a pas été vu. On l'ignore.
            log_prob += math.log(1e-10)  # Utilisation d'un très petit nombre pour éviter les problèmes

    # Normalisation par le nombre de tokens (en minuscules) dans le texte
    perplexity = math.exp(-log_prob / len(bigram_list))
    return perplexity

def preprocess_for_topic_modeling(abstracts):
    """Prépare les abstracts pour le modèle de topic modeling."""
    # Stopwords en anglais
    stop_words = set(stopwords.words('english'))
    
    # Liste pour stocker les textes nettoyés
    texts = []
    
    for abstract in abstracts:
        # Tokenisation avec spaCy
        doc = nlp(abstract.lower())
        
        # Filtrer les tokens : enlever les stopwords et la ponctuation
        tokens = [token.lemma_ for token in doc if token.is_alpha and token.text not in stop_words]
        
        texts.append(tokens)
    
    return texts

def perform_topic_modeling(texts, num_topics=3):
    """Effectue le topic modeling avec LDA."""
    # Créer un dictionnaire et un corpus
    dictionary = corpora.Dictionary(texts)
    corpus = [dictionary.doc2bow(text) for text in texts]
    
    # Appliquer LDA (Latent Dirichlet Allocation)
    lda_model = gensim.models.LdaMulticore(corpus, num_topics=num_topics, id2word=dictionary, passes=15, workers=2)
    
    return lda_model

# Spécifie le chemin du fichier XML
xml_path = "C:/Users/user/Downloads/sample-0001.xml.gz"
xml_file = decompress_gz(xml_path)

# Extraire les 10 premiers abstracts
abstracts = extract_abstracts(xml_file, limit=10)

# Préparation des données pour le topic modeling
texts = preprocess_for_topic_modeling(abstracts)

# Appliquer le Topic Modeling
lda_model = perform_topic_modeling(texts, num_topics=3)

# Affichage des trois topics les plus cités
print("\n--- Les trois topics les plus cités ---")
for idx, topic in lda_model.print_topics(num_topics=3, num_words=10):
    print(f"Topic {idx + 1}: {topic}")
```
